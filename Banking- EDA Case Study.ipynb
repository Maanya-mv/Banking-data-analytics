```python
bins=[0,100000,300000,float('inf')]
labels=['Low', 'Med', 'High']

df['Income Band']=pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
#Examine the distribution of unique categories in categorical columns

categorical_cols = df[['BRId', 'GenderId', 'IAId', 'Amount of Credit Cards', 'Nationality', 
                       'Occupation', 'Fee Structure', 'Loyalty Classification', 
                       'Properties Owned', 'Risk Weighting', 'Income Band']]

for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts())
```

```python
#Univariate Analysis
for i, col in enumerate(['BRId', 'GenderId', 'IAId', 'Amount of Credit Cards', 'Nationality', 
                         'Occupation', 'Fee Structure', 'Loyalty Classification', 
                         'Properties Owned', 'Risk Weighting', 'Income Band']):
    plt.figure(i)
    sns.countplot(data=df, x=col)
    plt.show()
```

```python
#Bivariate Analysis
for i, col in enumerate(['BRId', 'GenderId', 'IAId', 'Amount of Credit Cards', 'Nationality', 
                         'Occupation', 'Fee Structure', 'Loyalty Classification', 
                         'Properties Owned', 'Risk Weighting', 'Income Band']):
    plt.figure(i)
    sns.countplot(data=df, x=col, hue='GenderId')
    plt.show()
```

```python
#Bivariate Analysis
for i, col in enumerate(['BRId', 'GenderId', 'IAId', 'Amount of Credit Cards', 'Nationality', 
                         'Occupation', 'Fee Structure', 'Loyalty Classification', 
                         'Properties Owned', 'Risk Weighting', 'Income Band']):
    plt.figure(i)
    sns.countplot(data=df, x=col, hue='Nationality')
    plt.show()
```

```python
# Histplot of value counts (excluding 'Occupation')

for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col])  
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
#Numerical Analysis

numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans',
               'Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account',
               'Business Lending']

#Univariate Analysis and visualization
plt.figure(figsize=(50,50))#change accordingly
for i,col in enumerate(numerical_cols):
    plt.subplot(9,1,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()    
    
    
```

```python
#Heatmaps

correlation_matrix=df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest',fmt='.2f')
plt.show()
```

```python
#Insights of EDA

1.The strongest positive correlation occur among "Bank deposits" with "Checking
Accounts","Saving Accounts" and "Foreign Currency Account" indicating that customers
who maintain high balances in one account type often hold substantial amount/funds
across other accounts as well
```
